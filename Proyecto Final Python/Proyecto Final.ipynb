<p>4.1.1. ¿Cuál es el ingreso anual de las propiedades vendidas?

```python
# Agrupar variables
dfTemp1= df_bienes_raices.groupby('listYear')['saleAmount'].sum()

# Definir el formato de visualización con separador de millares
def format_with_commas(value):
    return '{:,}'.format(value)

# Aplicar el formato a la Serie
dfTemp1= dfTemp1.apply(format_with_commas)

# Imprimir el DataFrame
print(dfTemp1)
```

<p> El resultado muestra que el año 2020 fue en el que se registro el mayor ingreso por concepto de propiedades vendidas, seguido en orden descendente por el año 2002, 2001 y 2003.</p>

<p>4.1.2. ¿Qué tipo de propiedad tiene mayor facilidad de venta?</p>

```python
dfTemp2= df_bienes_raices.groupby(["propertyType"]).count()
dfTemp2= dfTemp2.iloc[:,:1]
dfTemp2= dfTemp2.sort_values('serialNumber', ascending=False)
dfTemp2
```

<p> El resultado muestra que el tipo de propiedad con mayor facilidad de venta es de tipo Residencial, no obstante se debe ser cauteloso con esta información dado que debe considerarse que en el análisis se mantuvieron más del 60% de valores nulos que componen el DataFrame. </p>

<p>4.1.3.¿En cuál ciudad o pueblo se dan mayor cantidad de ventas de propiedades? Y ¿Qué elementos particulares tienen esas ciudades o pueblos? <p>

<p>4.1.4.¿Cuál es el costo promedio de las propiedades residenciales y comerciales? <p>

```python
costo_promedio_residencial = df_bienes_raices[df_bienes_raices['propertyType'] == 'Residential']['saleAmount'].mean()
costo_promedio_comercial = df_bienes_raices[df_bienes_raices['propertyType'] == 'Commercial']['saleAmount'].mean()

# Formatear los costos promedio con separadores de millares
costo_promedio_residencial= "{:,.2f}".format(costo_promedio_residencial)
costo_promedio_comercial= "{:,.2f}".format(costo_promedio_comercial)

print("Costo promedio de propiedades residenciales:", costo_promedio_residencial)
print("Costo promedio de propiedades comerciales:", costo_promedio_comercial)
```

<h4 Style= 'color: #f08309' >Configuración de gráfico</h4> 

```python
#Remueve las lineas de los axis.
mpl.rcParams["axes.spines.bottom"] = True
mpl.rcParams["axes.spines.left"] = True
mpl.rcParams["axes.spines.right"] = False
mpl.rcParams["axes.spines.top"] = False
```

```python
# Datos obtenidos
costo_promedio_residencial = 479441.76
costo_promedio_comercial = 1771751.58

# Etiquetas para las barras
etiquetas = ['Residencial', 'Comercial']

# Valores para las barras
valores = [costo_promedio_residencial, costo_promedio_comercial]

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(8, 6))

bar_colors = [ '#e5e5e5','#69d2e7']
        
ax.bar(etiquetas, valores, color=bar_colors)


#--------------------------------------------------------------------------------------------#
#-----------------CONFIGURACION DE LOS TITULOS, SUBTITULOS Y EFES DEL GRAFICO----------------#
#--------------------------------------------------------------------------------------------#

#FORMATO DEL TITULO

# Configuraciones del gráfico

title = "Gráfico 1: Costo Promedio de Propiedades en Connecticut"
plt.suptitle(title,
             horizontalalignment = 'left',
             x = 0.001,
             y = 0.87,
             transform = fig.transFigure,
             fontsize = 20,
             fontweight = "bold",
             fontfamily = "serif",
             fontname = "Liberation Sans Narrow")

#FORMATO DEL SUBTITULO
subTitle = "Costo Promedio de Propiedades Residenciales y Comerciales en dólares."
ax.set_title(subTitle,
             horizontalalignment = "left",
             x = 0.001,
             y = 0.79,
             transform = fig.transFigure,
             fontsize = 15,
             fontweight = "regular",
             fontfamily = "serif")

#FORMATO AL EJE (X/Y)
etiqueta = "Tipo de Propiedad"
ax.set_xlabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

etiqueta = "Costo Promedio"
ax.set_ylabel(etiqueta,
              ha = 'left',
              fontsize = 10,
              fontweight = "regular",
              fontfamily = "serif")

ax.xaxis.set_label_coords(x=0, y=-0.07, transform = ax.transAxes)

#FORMATO TAMAÑO LETRA AMBOS EJES
ax.tick_params(axis='both', labelsize = 10)
 

#CONFIGURA LA UBICACION DEL TITULO, SUBTITULO Y GRAFICO
plt.subplots_adjust(top = 0.75)

# MOSTRAR EL VALOR DE CADA BARRA ENCIMA DE ELLAS
for i, v in enumerate(valores):
    ax.text(i, v + 20000, "${:,.2f}".format(v), ha='center', fontsize=12)

# MOSTRAR EL GRÁFICO
plt.show()
```

<p> El resultado muestra que las propiedades Comerciales tienden a tener un costo promedio mucho mayor de aproximadamente 1,771,751.58 en comparación a las propiedades Residenciales con un costo promedio aproximado de $479,441.76</p>

4.1.5.¿Cuál es el tipo de propiedad residencial que se vende con mayor frecuencia, donde y con que costo promedio?

```python
# Tipo de propiedad residencial que se vende con mayor frecuencia
dfTemp4= df_bienes_raices.groupby(["residentialType"]).count()
dfTemp4= dfTemp4.iloc[:,:1]
dfTemp4= dfTemp4.sort_values('serialNumber', ascending=False)
dfTemp4
```

```python
# Donde se vende con mayor frecuencia la propiedad residencial.
tipo_propiedad_frecuente = df_bienes_raices[df_bienes_raices['propertyType'] == 'Residential']['residentialType'].value_counts().idxmax()
ubicacion_tipo_frecuente = df_bienes_raices[(df_bienes_raices['propertyType'] == 'Residential') & (df_bienes_raices['residentialType'] == tipo_propiedad_frecuente)]['town'].value_counts().idxmax()

# Costo promedio de la propiedad residencial que se vende con mayor frecuencia.
costo_promedio_tipo_frecuente = df_bienes_raices[(df_bienes_raices['propertyType'] == 'Residential') & (df_bienes_raices['residentialType'] == tipo_propiedad_frecuente)]['saleAmount'].mean()

# Formatear los costos promedio con separadores de millares
costo_promedio_tipo_frecuente= "{:,.2f}".format(costo_promedio_tipo_frecuente) 

print("Tipo de propiedad residencial más frecuente:", tipo_propiedad_frecuente)
print("Ubicación donde se vende más el tipo de propiedad:", ubicacion_tipo_frecuente)
print("Costo promedio del tipo de propiedad más frecuente:", costo_promedio_tipo_frecuente)
```

<p> El resultado muestra que la propiedad residencial que se vende con mayor frecuencia es de tipo Single Family, con un costo promedio de $514,531.54 y ubicada en el pueblo de Greenwich. </p>

4.1.6.¿Cuál es el margen promedio de ganancia por cada una de las propiedades vendidas?

```python
# Calcular el sales ratio 
df_bienes_raices['salesRatio'] = df_bienes_raices['saleAmount'] / df_bienes_raices['assessedValue']

df_bienes_raices. head(5)
```

```python
# Filtrar filas donde el valor de 'assessedValue' es cero
filtered_df = df_bienes_raices[df_bienes_raices['assessedValue'] != 0]

# Calcular el promedio de la columna salesRatio
sales_ratio_mean = filtered_df['salesRatio'].mean()

# Imprimir el resultado
print("Margen Promedio de Ganancia:", sales_ratio_mean)
```

<p> El resultado muestra que el margen promedio de ganancia por cada una de las propiedades vendidas es de aproximadamente 5.4% </p>
